```
import os
from google.colab import files
from langchain.document_loaders import DirectoryLoader, TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA
from transformers import pipeline
```

```
os.makedirs("knowledge-base/employees", exist_ok=True)
```

```
uploaded = files.upload()
for filename in uploaded.keys():
    with open(f"knowledge-base/employees/{filename}", "wb") as f:
        f.write(uploaded[filename])
```

```
DOCS_PATH = "knowledge-base/employees"
loader = DirectoryLoader(DOCS_PATH, glob="**/*.md", loader_cls=TextLoader)
documents = loader.load()
print(f"Loaded {len(documents)} documents.")
```

```
print(documents)
```

```
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = text_splitter.split_documents(documents)
print(f"Split into {len(chunks)} chunks.")
```

```
print(chunks)
```

```
embedding_model = "sentence-transformers/all-MiniLM-L6-v2"  # Lightweight & fast
embedding = HuggingFaceEmbeddings(model_name=embedding_model)
print("Embedding model loaded successfully!")
```

```
vectorstore = FAISS.from_documents(chunks, embedding)
retriever = vectorstore.as_retriever()
print("FAISS vector store created")
```

```
llm_pipeline = pipeline("text-generation", model="mosaicml/mpt-7b", device_map="auto",max_new_tokens=50)
llm = HuggingFacePipeline(pipeline=llm_pipeline)
print("Hugging Face LLM loaded successfully!")
```

```
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever
)
print("QA System ready!")
```

```
def chat(query: str) -> str:
    retrieved_context = qa_chain.invoke(query)

    if not retrieved_context:
        return "I don't know."

    prompt = f"""
    Use ONLY the following retrieved context to answer the question.
    If you don't know the answer, say "I don't know."

    Context:
    {retrieved_context}

    Question: {query}
    Answer:
    """

    response = llm_pipeline(prompt, max_new_tokens=50)

    return response[0]["generated_text"].strip()

print(chat("Who is Alex?"))
```
